# Saved Figures

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Generate example data

```python
#Generar datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

# Create the figure and the axis

```python
#Crear la figura y el eje
fig, ax = plt.subplots(figsize=(8, 5))

#Dibujar la función
ax.plot(x, y, label='y = sin(x)', color='blue', linestyle='-', linewidth=1)
ax.set_title('Función Seno')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.legend()
ax.grid(True, linestyle='--', linewidth=0.6)

#Guardar en diferentes formatos
plt.savefig('grafico.png', dpi=300)
plt.savefig('grafico.pdf')
plt.savefig('grafico.svg')
plt.savefig('grafico.jpg')

#Mostrar el gráfico
plt.show()
```

### Explanation
### Guardar en diferentes formatos
- ``plt.savefig("nombre.png", dpi=300)`` -> Guarda en PNG con alta calidad
- ``plt.savefig("nombre.pdf")`` -> Guarda en PDF (ideal para impresión)
- ``plt.savefig("nombre.svg")`` -> Guarda en SVG (para graficos vectoriales escalables)
- ``plt.savefig("nombre.jpg")`` -> Guarda en JPG con compresión

### Parametros utiles de ``savefig()``
- ``dpi=300`` -> Control la resolución (DPI = puntos por pulgada)
- ``quality=95`` -> Para imágenes JPG, ajusta la calidad (0-100)
- ``transparent=True`` -> Guarda el fondo transparente si es compatible
