# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Generate example data

```python
#Generar datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2 * x)
y4 = np.cos(2 * x)
```

## Create the figure and subgraphs (2 rows, 2 columns)

```python
#Crear la figura y los subgrficos (2 filas, 2 columnas)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

#Subplot 1: Seno
axs[0, 0].plot(x, y1, color="blue", linestyle="--", label="Seno")
axs[0, 0].set_title("Seno")
axs[0, 0].legend()
axs[0, 0].grid(True)

#Subplot 2: Coseno
axs[0, 1].plot(x, y2, color="red", linestyle="-.", label="Coseno")
axs[0, 1].set_title("Coseno")
axs[0, 1].legend()
axs[0, 1].grid(True)

#Subplot 3: Seno con frecuencia doble
axs[1, 0].plot(x, y3, color="green", linestyle=":", label="Seno(2x)")
axs[1, 0].set_title("Seno(2x)")
axs[1, 0].legend()
axs[1, 0].grid(True)

#Subplot 4: Coseno con frecuencia doble
axs[1, 1].plot(x, y4, color="purple", linestyle="-", label="Coseno(2x)")
axs[1, 1].set_title("Coseno(2x)")
axs[1, 1].legend()
axs[1, 1].grid(True)

#Ajustar el layout para evitar solapamientos
plt.tight_layout()

#Mostrar los grficos
plt.show()
```

### Explanation
### Crear una figura con mltiples subgrficos
- ``fig, axs = plt.subplots(2, 2, figsize=(10, 8))`` -> Crea una figura con 2 filas y 2 colunas de subgrficos
- ``axs`` es una matriz de 2x2 para acceder a cada subplot

### Asignar datos a cada subplot
- ``axs[fila, columna]`` -> Se utiliza para acceder a cada subgrfico
- ``set_title()``, ``legend()`` y ``grid()`` -> Personalizan cada subplot

### Ajustar el diseo automaticamente
- ``plt.tight_layout()`` -> Evita que los ttulos y etiquetas se superpongan
