# Line Graph

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Example data

```python
#Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

## Create a graph and axis

```python
#Crear una grafica y un eje
fig, ax = plt.subplots()

#Graficar dos lineas con distintos estilos
ax.plot(x, y1, label="Seno", color="blue", linestyle="-", linewidth=2, marker="o", markersize=4)
ax.plot(x, y2, label="Coseno", color="red", linestyle="--", linewidth=2, marker="s", markersize=4)

#Personalizar la grafica
ax.set_title("Grafico de lineas con Estilos")
ax.set_xlabel("Eje x")
ax.set_ylabel("Eje y")
ax.legend() #Agregar leyenda
plt.grid(True) #Agregar cuadricula

#Mostrar la grafica
plt.show()
```

### Explanation
- ``linestyle="-"`` -> Linea sólida, también puedes usar ``"--", "-.", ":"``
- ``linewidth=2`` -> Grosor de linea
- ``marker="o"`` -> Marcadores een cada punto (``"s"`` cuadrado, ``"^"`` triangulo, ``"x"`` cruz, etc.)
- ``ax.legend()`` -> Agrega leyenda automatica
- ``ax.grid(True)`` -> Activa la cuadricula
