# Bars and Pies

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Data for both graphs

```python
#Datos para ambos graficos
categorys = ['A', 'B', 'C', 'D', 'E']
values = [7, 13, 5, 17, 20]
```

## Create the figure and the axes

```python
#Crear la figura y los ejes
fig, axs = plt.subplots(1, 2, figsize=(12, 6)) #Un grafico de barras y otro de pastel

##Grafico de barras
axs[0].bar(categorys, values, color='skyblue', edgecolor="black")

#Personalizacion
axs[0].set_title("Grafico de Barras")
axs[0].set_xlabel("Categoria")
axs[0].set_ylabel("Valores")
axs[0].grid(axis="y", linestyle="--", alpha=0.7)

#Grafico de pastel
axs[1].pie(values, labels=categorys, autopct="%1.1f%%", startangle=90, colors=["blue", "red", "green", "yellow", "purple"], shadow=True)

#Personalizacion
axs[1].set_title("Grafico de Pastel")

#Ajustar el layout para evitar sobreposicion
plt.tight_layout()

#Mostrar el grafico
plt.show()
```

### Explanation
### Grafico de Barras:
- ``axs[0].bar(categorys, values, color="skyblue", edgecolor="black")`` -> Crea las barras con borde negro
- ``axs[0].grid(axis="y", linestyle="--", alpha=0.7)`` -> Agrega una cuadricula en el eje Y

### Grafico de Pastel:
- ``axs[1].pie(valores, labels=categorias, autopct="%1.1f%%", startangle=90, shadow=True)``
- ``autopct="%1.1f%%"`` -> Muestra los porcentajes en el grafico.
- ``startangle=90`` -> Rota el inicio del grafico en 90°
- ``shadow=True`` -> Agrega sombra al grafico
