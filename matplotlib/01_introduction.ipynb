# Introduction

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Example data

```python
#Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

## Create a figure and axis

```python
#Crear una figura y un eje
fig, ax = plt.subplots()
ax.plot(x, y)

#Titulo y etiquetas
ax.set_title('Grafico de Linea Básico')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')

#Mostrar el gráfico
plt.show()
```
