# Bars and Pies

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Data for both graphs

```python
#Datos para ambos gr√°ficos
categorys = ['A', 'B', 'C', 'D', 'E']
values = [7, 13, 5, 17, 20]
```

## Create the figure and the axes

```python
#Crear la figura y los ejes
fig, axs = plt.subplots(1, 2, figsize=(12, 6)) #Un gr√°fico de barras y otro de pastel

##Gr√°fico de barras
axs[0].bar(categorys, values, color='skyblue', edgecolor="black")

#Personalizaci√≥n
axs[0].set_title("Gr√°fico de Barras")
axs[0].set_xlabel("Categor√≠a")
axs[0].set_ylabel("Valores")
axs[0].grid(axis="y", linestyle="--", alpha=0.7)

#Gr√°fico de pastel
axs[1].pie(values, labels=categorys, autopct="%1.1f%%", startangle=90, colors=["blue", "red", "green", "yellow", "purple"], shadow=True)

#Personalizaci√≥n
axs[1].set_title("Gr√°fico de Pastel")

#Ajustar el layout para evitar sobreposici√≥n
plt.tight_layout()

#Mostrar el gr√°fico
plt.show()
```

### Explanation
### Gr√°fico de Barras:
- ``axs[0].bar(categorys, values, color="skyblue", edgecolor="black")`` -> Crea las barras con borde negro
- ``axs[0].grid(axis="y", linestyle="--", alpha=0.7)`` -> Agrega una cuadr√≠cula en el eje Y

### Gr√°fico de Pastel:
- ``axs[1].pie(valores, labels=categorias, autopct="%1.1f%%", startangle=90, shadow=True)``
    - ``autopct="%1.1f%%"`` -> Muestra los porcentajes en el gr√°fico
    - ``startangle=90`` -> Rota el inicio del gr√°fico en 90¬∞
    - ``shadow=True`` -> Agrega sombra al gr√°fico
