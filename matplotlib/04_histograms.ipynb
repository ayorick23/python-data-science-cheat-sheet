# Histograms

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Generate random data with normal distribution

```python
#Generar datos aleatorios con distribucion normal
np.random.seed(42) #para reproductibilidad
data = np.random.randn(1000) #1000 valores con distribuci√≥n normal
```

## Create the figure and the axis

```python
#Crear la figura y el eje
fig, ax = plt.subplots(figsize=(8, 6))

#Crear histograma
ax.hist(data, bins=30, color="royalblue", edgecolor="black", alpha=0.7)

#Personalizaci√≥n
ax.set_title("Histograma de Distribuci√≥n Normal")
ax.set_xlabel("Valores")
ax.set_ylabel("Frecuencia")
ax.grid(axis="y", linestyle="--", alpha=0.7)

#Mostrar la gr√°fica
plt.show()
```

### Explanation
- ``np.random.randn(1000)`` -> Genera 1000 valores con distribuci√≥n normal est√°ndar (media = 0, desviaci√≥n est√°ndar = 1)
- ``bins=30`` -> N√∫mero de barras en el histograma
- ``color="royalblue"`` -> Color de las barras
- ``edgecolor="black"`` -> Borde de las barras
- ``alpha=0.7`` -> Transparencia
- ``ax.grid(axis="y", linestyle="--", alpha=0.7)`` -> Activa la cuadricula en el eje Y
