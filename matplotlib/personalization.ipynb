# Personalization

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Generate example data

```python
#Generar datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
#Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

#Agregar dos lineas con diferentes estilos
ax.plot(x, y1, label="Seno", color="blue", linestyle="--", linewidth=2, marker="o", markersize=5, markerfacecolor="red")
ax.plot(x, y2, label="Coseno", color="green", linestyle="-.", linewidth=2, marker="s", markersize=5, markerfacecolor="orange")

#Personalizacion del grafico
ax.set_title("Grafico de Seno y Coseno", fontsize=14, fontweight="bold")
ax.set_xlabel("Eje X", fontsize=12)
ax.set_ylabel("Eje Y", fontsize=12)
ax.legend(loc="upper right", fontsize=12)
ax.grid(True, linestyle="--", alpha=0.6)

#Personalizacion de ejes
ax.set_xticks(np.arange(0, 11, 1)) #marcas en el eje x cada 1 unidad
ax.set_yticks(np.arange(-1, 1.2, 0.2)) #marcas en el eje y cada 0.2 unidades

#Cambiar el color de fondo
ax.set_facecolor("#f5f5f5")

#Mostrar el grafico
plt.show()
```

### Explanation
### Personalizacion de lineas
- ``linestyle="--"`` -> Linea punteada par seno
- ``linestyle="-."`` -> Linea con puntos y guiones para coseno
- ``linewidth=2`` -> Grosor de las lineas
- ``marker="o"`` y ``marker="s"`` -> Marcadores (circulos y cuadrados)
- ``markerfacecolor="red"`` y ``"orange"`` -> Color de los marcadores.

### Personalizacion de titulos y etiquetas
- ``fontsize=14, fontweight="bold"`` -> Tamao y negrita en el titulo

### Leyenda y grid
- ``ax.legend(loc="upper right")`` -> Muestra la leyenda en la esquina superior derecha
- ``ax.grid(True, linestyle="--", alpha=0.6)`` -> Cuadricula con lineas punteadas y transparencia

### Configuracion de ticks (marcas en los ejes)
- ``ax.set_xticks(np.arange(0, 11, 1))`` -> Marca en X cada 1 unidad
- ``ax.set_yticks(np.arange(-1, 1.2, 0.2))`` -> Marca en Y cada 0.2 unidades

### Color de fondo
- ``ax.set_facecolor("#f5f5f5")`` -> Fondo gris claro
