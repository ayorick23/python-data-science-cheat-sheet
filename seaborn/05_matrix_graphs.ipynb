# Matrix Graphs

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
```

```python
#Cargar dataset de ejemplo
iris = sns.load_dataset("iris")
```

```python
#Configurar estilo
sns.set_style("whitegrid")
```

```python
#Seleccionar solo columnas num√©ricas
iris_numeric = iris.select_dtypes(include=['float64', 'int64'])
```

```python
#Gr√°fico de correlaci√≥n
plt.figure(figsize=(8, 6))
sns.heatmap(iris_numeric.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Mapa de Calor de Correlaci√≥n entre Variables")
plt.show()
```

```python
#Matriz de dispersi√≥n con pairplot
sns.pairplot(iris, hue="species", diag_kind="kde")
plt.suptitle("Matriz de dispersi√≥n: Relaci√≥n entre Variables")
plt.show()
```

```python
#Clustermap para visualizar agrupaciones
sns.clustermap(iris.drop(columns=["species"]), cmap="coolwarm", standard_scale=1)
plt.title("Clustermap de las Variables Num√©ricas")
plt.show()
```
