# Style Configuration

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Datos de ejemplo
tips = sns.load_dataset("tips")
```

```python
#Configurar estilo
sns.set_style("darkgrid")
```

```python
#Crear un grfico de dispersin con estilo configurado
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")

#Ttulo y etiquetas
plt.title("Relacin entre Total de la Cuenta y Propina")
plt.xlabel("Total de la Cuenta")
plt.ylabel("Propina")

#Mostrar grfico
plt.show()
```

### Explanation
### Principales estilos en Seaborn
- ``darkgrid`` -> Fondo gris oscuro con cuadrcula (por defecto en Seaborn)
- ``whitegrid`` -> Fondo blanco con cuadrcula
- ``dark`` -> Fondo gris oscuro sin cuadrcula
- ``white`` -> Fondo blanco sin cuadrcula
- ``ticks`` -> Fondo blanco con lneas en los ejes

### Paletas de colores en Seaborn
- ``deep`` -> Colores variados y equilibrados
- ``muted`` -> Tonos suaves
- ``bright`` -> Colores vivos y saturados
- ``dark`` -> Tonos oscuros
- ``colorblind`` -> Optimizado para personas con daltonismo
