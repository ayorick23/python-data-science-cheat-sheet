# Introduction

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load sample dataset

```python
#Cargar dataset de ejemplo
df = sns.load_dataset("penguins")
```

## Create a basic graphic with Seaborn

```python
#Crear un gráfico básico con Seaborn
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species")

#Título del gráfico
plt.title("Relación entre Longitud y Profundidad del Pico de los Pingüinos")

#Mostrar gráfico
plt.show()
```

### Explanation
- ``data=df`` -> Se especifica el **DataFrame** que contiene los datos
- ``x="bill_length_mm"`` -> Se elige la columna que estará en el eje X
- ``Y="bill_depth_mm"`` -> Se elige la columna para el eje Y
- ``hue="species"`` -> Se asigna un color distintno a cada categoría de la columna **species**

### Otros argumentos importantes
| Argumento | Descripción | Ejemplo |
|-----------|-------------|---------|
| ``size`` | Ajusta el tamaño de los puntos según una variable numérica | ``size="body_mass_g"`` |
| ``style`` | Cambia el estilo de los puntos según una variable categórica | ``style="island"`` |
| ``palette`` | Cambia los colores de las categorías | ``palette="coolwarm"`` |
| ``alpha`` | Controla la transparencia de los puntos *(0 = invisible, 1 = opaco)* | ``alpha=0.7`` |
| ``edgecolor`` | Define el color del borde de los puntos | ``edgecolor="black"`` |
| ``linewidth`` | Grosor del borde de los puntos | ``linewidth=1.5``  |
