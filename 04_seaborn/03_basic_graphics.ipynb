# Basic Graphics

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load sample dataset

```python
#Cargar dataset de ejemplo
tips = sns.load_dataset("tips")
```

## Set style

```python
#Configurar estilo
sns.set_style("whitegrid")
```

## Scatter plot

```python
#Gráfico de dispersión
plt.figure(figsize=(12, 6))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.title("Gráfico de Dispersión: Total de la Cuenta vs Propina")
plt.show()
```

## Line plot

```python
#Gráfico de líneas
plt.figure(figsize=(12, 6))
sns.lineplot(data=tips, x="size", y="total_bill", hue="day", marker="o")
plt.title("Gráfico de Líneas: Relación entre Tamaño de Grupo y Total de la Cuenta")
plt.show()
```

## Bar plot

```python
#Gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(data=tips, x="day", y="total_bill", hue="sex", ci=None)
plt.title("Gráfico de Barras: Promedio del Total de la Cuenta por Día y Género")
plt.show()
```

## Box plot

```python
#Gráfico de caja
plt.figure(figsize=(12, 6))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Gráfico de Caja: Distribución del Total de la Cuenta")
plt.show()
```

## Violin plot

```python
#Gráfico de violin
plt.figure(figsize=(12, 6))
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.title("Gráfico de Violín: Distribución del Total de la Cuenta por Día y Género")
plt.show()
```

## Point plot

```python
#Gráfico de punto
plt.figure(figsize=(12, 6))
sns.pointplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.title("Gráfico de Puntos: Tendencias del Total de la Cuenta por Día y Género")
plt.show()
```

## Swarm plot

```python
#Gráfico de enjambre
plt.figure(figsize=(12, 6))
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.title("Gráfico de Enjambre: Distribución de Propinas")
plt.show()
```

### Explanation
- ``hue`` -> Diferenciar datos por color
- ``ci`` -> Control de barras de error en gráficos de barras
    - ``ci=None`` -> No muestra barras de error
    - ``ci=95`` -> Intervalo de confianza del 95%
- ``split`` -> Separar por categoría
- ``dodge`` -> Separa puntos de diferentes categorías en ``pointplot()`` y ``swarmplot()``
