# Basic Graphics

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load sample dataset

```python
#Cargar dataset de ejemplo
tips = sns.load_dataset("tips")
```

## Set style

```python
#Configurar estilo
sns.set_style("whitegrid")
```

## Scatter plot

```python
#Grfico de dispersin
plt.figure(figsize=(12, 6))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.title("Grfico de Dispersin: Total de la Cuenta vs Propina")
plt.show()
```

## Line plot

```python
#Grfico de lneas
plt.figure(figsize=(12, 6))
sns.lineplot(data=tips, x="size", y="total_bill", hue="day", marker="o")
plt.title("Grfico de Lneas: Relacin entre Tamao de Grupo y Total de la Cuenta")
plt.show()
```

## Bar plot

```python
#Grfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(data=tips, x="day", y="total_bill", hue="sex", ci=None)
plt.title("Grfico de Barras: Promedio del Total de la Cuenta por Da y Gnero")
plt.show()
```

## Box plot

```python
#Grfico de caja
plt.figure(figsize=(12, 6))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Grfico de Caja: Distribucin del Total de la Cuenta")
plt.show()
```

## Violin plot

```python
#Grfico de violin
plt.figure(figsize=(12, 6))
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.title("Grfico de Violn: Distribucin del Total de la Cuenta por Da y Gnero")
plt.show()
```

## Point plot

```python
#Grfico de punto
plt.figure(figsize=(12, 6))
sns.pointplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.title("Grfico de Puntos: Tendencias del Total de la Cuenta por Da y Gnero")
plt.show()
```

## Swarm plot

```python
#Grfico de enjambre
plt.figure(figsize=(12, 6))
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.title("Grfico de Enjambre: Distribucin de Propinas")
plt.show()
```

### Explanation
- ``hue`` -> Diferenciar datos por color
- ``ci`` -> Control de barras de error en grficos de barras
    - ``ci=None`` -> No muestra barras de error
    - ``ci=95`` -> Intervalo de confianza del 95%
- ``split`` -> Separar por categora
- ``dodge`` -> Separa puntos de diferentes categoras en ``pointplot()`` y ``swarmplot()``
