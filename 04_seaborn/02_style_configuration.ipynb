# Style Configuration

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Datos de ejemplo
tips = sns.load_dataset("tips")
```

```python
#Configurar estilo
sns.set_style("darkgrid")
```

```python
#Crear un gráfico de dispersión con estilo configurado
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")

#Título y etiquetas
plt.title("Relación entre Total de la Cuenta y Propina")
plt.xlabel("Total de la Cuenta")
plt.ylabel("Propina")

#Mostrar gráfico
plt.show()
```

### Explanation
### Principales estilos en Seaborn
- ``darkgrid`` -> Fondo gris oscuro con cuadrícula (por defecto en Seaborn)
- ``whitegrid`` -> Fondo blanco con cuadrícula
- ``dark`` -> Fondo gris oscuro sin cuadrícula
- ``white`` -> Fondo blanco sin cuadrícula
- ``ticks`` -> Fondo blanco con líneas en los ejes

### Paletas de colores en Seaborn
- ``deep`` -> Colores variados y equilibrados
- ``muted`` -> Tonos suaves
- ``bright`` -> Colores vivos y saturados
- ``dark`` -> Tonos oscuros
- ``colorblind`` -> Optimizado para personas con daltonismo
