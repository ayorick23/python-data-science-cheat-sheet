# Scatter Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Generate random data

```python
#Generar datos aleatorios
np.random.seed(42) #para que los datos sean los mismos en cada ejecución (reproductividad)
x = np.random.rand(100) * 10 #100 datos aleatorios entre 0 y 10
y = np.random.rand(100) * 10
sizes = np.random.rand(100) * 300 #tamaño aleatorio de los puntos
colors = np.random.rand(100) #color aleatorio de los puntos
```

## Create the figure and the axis

```python
#Crear la figura y el eje
fig, ax = plt.subplots(figsize=(8, 6)) #tamaño de la figura

#Gráfico de dispersion
scatter = ax.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7, edgecolors="k", linewidth=2) #s=tamaño, c=color, cmap=colormap, alpha=transparencia, edgecolors=bordes, linewidth=grosor de los bordes

#Personalización
ax.set_title("Gráfico de Dispersión", fontsize=15)
ax.set_xlabel("Eje x", fontsize=12)
ax.set_ylabel("Eje y", fontsize=12)
ax.grid(True)

#Agregar barra de colores
cbar = plt.colorbar(scatter)
cbar.set_label("Intensidad del color")

#Mostrar la gráfica
plt.show()
```

### Explanation
- ``s=sizes`` -> Tamaño de los puntos
- ``c=colors`` -> Color de los puntos
- ``cmap="viridis"`` -> Mapa de colores (puedes usar ``"plasma"``, ``"coolwarm"``, etc.)
- ``alpha=0.7`` -> Transparencia de los puntos
- ``edgecolors="k"`` -> Borde negro en los puntos
- ``plt.colorbar(scatter)`` -> Agrega una barra de color
