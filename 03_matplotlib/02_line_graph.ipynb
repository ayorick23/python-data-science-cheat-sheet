# Line Graph

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Example data

```python
#Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

## Create the figure and the axis

```python
#Crear la figura y el eje
fig, ax = plt.subplots()

#Graficar dos lneas con distintos estilos
ax.plot(x, y1, label="Seno", color="blue", linestyle="-", linewidth=2, marker="o", markersize=4)
ax.plot(x, y2, label="Coseno", color="red", linestyle="--", linewidth=2, marker="s", markersize=4)

#Personalizar la grfica
ax.set_title("Grafico de Lneas con Estilos")
ax.set_xlabel("Eje x")
ax.set_ylabel("Eje y")
ax.legend() #Agregar leyenda
plt.grid(True) #Agregar cuadrcula

#Mostrar la grfica
plt.show()
```

### Explanation
- ``linestyle="-"`` -> Linea slida, tambin puedes usar ``"--", "-.", ":"``
- ``linewidth=2`` -> Grosor de linea
- ``marker="o"`` -> Marcadores en cada punto (``"s"`` cuadrado, ``"^"`` tringulo, ``"x"`` cruz, etc.)
- ``ax.legend()`` -> Agrega leyenda automtica
- ``ax.grid(True)`` -> Activa la cuadrcula
