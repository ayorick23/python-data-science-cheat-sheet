# Customization and Styles

##  Personalización y Estilos

En **Matplotlib**, la personalización se puede lograr mediante el uso de estilos predefinidos o creando estilos propios, así como modificando directamente parámetros de configuración. Esto permite ajustar elementos como colores, fuentes, ejes, títulos y leyendas, entre otros, para adaptar los gráficos al estilo deseado.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Generate example data

```python
#Generar datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
#Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

#Agregar dos líneas con diferentes estilos
ax.plot(x, y1, label="Seno", color="blue", linestyle="--", linewidth=2, marker="o", markersize=5, markerfacecolor="red")
ax.plot(x, y2, label="Coseno", color="green", linestyle="-.", linewidth=2, marker="s", markersize=5, markerfacecolor="orange")

#Personalización del gráfico
ax.set_title("Gráfico de Seno y Coseno", fontsize=14, fontweight="bold")
ax.set_xlabel("Eje X", fontsize=12)
ax.set_ylabel("Eje Y", fontsize=12)
ax.legend(loc="upper right", fontsize=12)
ax.grid(True, linestyle="--", alpha=0.6)

#Personalización de ejes
ax.set_xticks(np.arange(0, 11, 1)) #marcas en el eje x cada 1 unidad
ax.set_yticks(np.arange(-1, 1.2, 0.2)) #marcas en el eje y cada 0.2 unidades

#Cambiar el color de fondo
ax.set_facecolor("#f5f5f5")

#Mostrar el gráfico
plt.show()
```

### Explanation
### Personalización de líneas
- ``linestyle="--"`` -> Línea punteada para seno
- ``linestyle="-."`` -> Línea con puntos y guiones para coseno
- ``linewidth=2`` -> Grosor de las líneas
- ``marker="o"`` y ``marker="s"`` -> Marcadores (círculos y cuadrados)
- ``markerfacecolor="red"`` y ``"orange"`` -> Color de los marcadores

### Personalización de títulos y etiquetas
- ``fontsize=14, fontweight="bold"`` -> Tamaño y negrita en el título

### Leyenda y grid
- ``ax.legend(loc="upper right")`` -> Muestra la leyenda en la esquina superior derecha
- ``ax.grid(True, linestyle="--", alpha=0.6)`` -> Cuadrícula con líneas punteadas y transparencia

### Configuración de ticks (marcas en los ejes)
- ``ax.set_xticks(np.arange(0, 11, 1))`` -> Marca en X cada 1 unidad
- ``ax.set_yticks(np.arange(-1, 1.2, 0.2))`` -> Marca en Y cada 0.2 unidades

### Color de fondo
- ``ax.set_facecolor("#f5f5f5")`` -> Fondo gris claro
