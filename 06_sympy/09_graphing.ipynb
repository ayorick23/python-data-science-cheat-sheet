# Graphing

```python
import sympy as sp
import sympy.plotting as syp
```

## Define symbolic variable

```python
#Definir variable simb贸lica
x = sp.Symbol('x')
```

## Define functions to graph

```python
#Definir funciones a graficar
f1 = sp.sin(x)  #funci贸n seno
f2 = sp.exp(-x) #funci贸n exponencial decreciente
f3 = x**2 - 4   #parab贸lica
```

## Graph functions

```python
#Graficar funciones
syp.plot(f1, f2, f3, (x, -5, 5), legend=True, show=True)
```

### Explanation
- ``syp.plot(f, (x, a, b))`` -> Dibuja la funci贸n ``f`` en el intervalo ``[a, b]``
- Se pueden graficar m煤ltiples funciones separadas por comas
- ``legend=True`` -> Muestra la leyenda autom谩tica de las funciones
- ``show=True`` -> Muestra el gr谩fico al ejecutarlo
